```python
import requests 
def ModelViewer(pmmlURI, statxmlURI):
    payload = '{"modelSpecification":{"algorithm":null,"inputs":[{"type":"pmml","uri":"' + pmmlURI + '"},{"type":"statxml","uri":"' + statxmlURI + '"}]},"globalFormatOptions":{"style":"dsx","htmlAssetsRootUrl":"https://model-viewer-dev.stage1.mybluemix.net","requireJS":true}}'
    headers = {'content-type': 'application/json'}
    r = requests.post('https://model-viewer-dev.stage1.mybluemix.net/mdlvr/visualizationsByModelSpecificationForCanvasEx', data=payload, headers=headers)
    return r.content
```

```python
from IPython.core.display import display, HTML
modelViewerHTML = ModelViewer("https://model-viewer-dev.stage1.mybluemix.net/mdlvr/testdata/LinearEngine/LE_pmml.xml", "https://model-viewer-dev.stage1.mybluemix.net/mdlvr/testdata/LinearEngine/LE_statxml.xml")
display(HTML(modelViewerHTML))
```
